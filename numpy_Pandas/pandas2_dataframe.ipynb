```python
import pandas as pd

data = {
    "ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Riya", "Rahul", "Amit", "Priya", "Neha", "Ankit"],
    "Age": [22, 25, 24, 23, 26, 22],
    "Salary": [35000, 42000, 50000, 45000, 55000, 38000],
    "Department": ["AI", "HR", "Finance", "Marketing", "AI", "AI"]
}

df = pd.DataFrame(data)

print(df)
df=df.sort_values(by="Salary") #sort in ascending order
print(df)
df=df.sort_values(by="Salary",ascending=False)
print(df)
df=df.sort_values(by=["Department","Salary"], ascending=[True,False]) #if two employees belomg to same department then it sort by salary
print(df)
df=df.sort_values("Name")
print(df)
df=df.sort_index()
print(df)
print(df.nlargest(3,"Salary")) #top 3 highest salary
print(df.nsmallest(2,"Age")) #smallest 2 age values
df["SalaryRank"]=df["Salary"].rank() #assigns rank from ascending to descending salary
print(df)
df["SalaryRank"]=df["Salary"].rank(ascending=False) #assigns rank from descending salary
print(df)
df["SalaryRank"]=df["Salary"].rank(method="average") #by default 
print(df)
df["SalaryRank"]=df["Salary"].rank(method="dense") 
print(df) 
```

```python
#groupby-used to split the data into groups,apply one or more operations to each group,combine the result
import pandas as pd

data = {
    "Name": ["Riya", "Rahul", "Amit", "Priya", "Neha", "Ankit"],
    "Department": ["AI", "HR", "Finance", "Marketing", "AI", "AI"],
    "Age": [22, 25, 24, 23, 26, 22],
    "Salary": [35000, 42000, 50000, 45000, 55000, 38000]
}

df = pd.DataFrame(data)

print(df)
df.groupby("Department")["Salary"].mean()
df.groupby("Department")["Salary"].sum()
df.groupby("Department")["Salary"].count() #count non null value row
df.groupby("Department")["Salary"].size() #count all rows even null values row
df.groupby("Department")["Salary"].min() #show min salary in each department
df.groupby("Department")["Salary"].max() #highest salary  in each department
df.groupby("Department")["Salary"].median() 
df.groupby("Department")["Salary"].std() #for department with only one employee the result is nan
df.groupby("Department")["Salary"].agg(["sum","mean","min","max","count"])
df.groupby("Department").agg(AverageSalary=("Salary","mean"),HighestSalary=("Salary","max"),TotalEmployees=("Name","count"))
df["Gender"]=["F","M","M","F","F","M"]
df.groupby(["Department","Gender"])["Salary"].mean() #group by multiple columns
#by default grouped columns become index,to keep it normal
df.groupby("Department",as_index=False)["Salary"].mean()
```

```python
import pandas as pd

df = pd.DataFrame({
    "Employee": ["Riya","Rahul","Riya","Neha","Rahul","Amit","Neha","Amit"],
    "Department": ["AI","HR","AI","AI","HR","Finance","AI","Finance"],
    "Product": ["Laptop","Mouse","Keyboard","Laptop","Laptop","Mouse","Keyboard","Laptop"],
    "Sales": [50000,1000,2000,55000,45000,1200,2500,60000]
})

print(df)
pd.pivot_table(
    df,
    values="Sales",
    index="Department",
    aggfunc="sum"
)
pd.pivot_table(
    df,
    values="Sales",
    index=["Department","Employee"],
    aggfunc="sum"
)
pd.pivot_table(
    df,
    values="Sales",
    index="Department",
    aggfunc=["sum","mean","max"],
    fill_value=0
)
pd.pivot_table(
    df,
    values="Sales",
    index="Department",
    columns="Product",
    aggfunc="sum",
    fill_value=0
)
df["Quantity"]=[5,1,1,6,4,2,2,5]
pd.pivot_table(
    df,
    values=["Sales","Quantity"],
    index="Department",
    aggfunc="sum"
)
pd.pivot_table(
    df,
    values="Sales",
    index="Department",
    aggfunc="sum",
    margins=True #grand total
)

pivot=pd.pivot_table(
    df,
    values="Sales",
    index="Department",
    aggfunc="sum",
)
pivot[pivot["Sales"]>46000]
df.groupby("Department")["Sales"].sum() #pivot vs group by
#pivot used only when there are no duplicate combinations of index and columns-if an employee has two sales of same product
```
