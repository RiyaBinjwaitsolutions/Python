```python
numbers=np.arange(2,101)
is_prime=np.ones(len(numbers),dtype=bool)
for i in range(2,int(np.sqrt(100))+1):
    is_prime[(numbers%i==0) & (numbers!=i)] = False

primes=numbers[is_prime]
print(primes)
```

```python
arr1=np.arange(1,7).reshape(2,3)
arr2=np.arange(1,7).reshape(3,2)
arr3=np.matmul(arr1,arr2)
print(arr3)
print(arr1@arr2)
arr4=np.arange(1,5).reshape(2,2)
print(np.linalg.det(arr4))
arr4_inv=np.linalg.inv(arr4)

print(np.linalg.trace(arr4))
print(np.linalg.matrix_rank(arr4))
values,vectors=np.linalg.eig(arr4)
print(values)
print(vectors)
print(np.matmul(arr4,arr4_inv))
```
