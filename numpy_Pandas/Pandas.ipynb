```python
import pandas as pd
import numpy as np
```

```python
print(pd.__version__)
```

```python
#from list
s=pd.Series([1,2,3,4])
print(s)
```

```python
#custom index
s=pd.Series(
    [1,2,3,4],
     index=["A","B","C","D"]
)
print(s)
```

```python
#from tuples
s=pd.Series((2,3,4))
print(s)
```

```python
#from numpy_array
arr=np.array([1,2,3,4])
print(arr)
s=pd.Series(arr)
print(s)
```

```python
#from dictionary - make keys as index
student = {
    "riya":90,
    "shreya":99,
    "kashi":87
}

s=pd.Series(student)
print(s)
```

```python
#from scalar values
s=pd.Series(100)
print(s)
#if you want multiple
s=pd.Series(100,
           index=["A","B","C","D"])
print(s)
```

```python
#attribute in series
s=pd.Series([1,2,3,4,5],
            index=["A","B","C","D","E"],
            name="Numbers"
             )
# print(s)
print(s.values) #values - return values of only
print(type(s.values)) #type of values
print(s.index) #returns only index 
print(s.dtype) #returns only datatype of series
print(s.shape) # returns dimension -series in 1D so shape contain one value
print(s.size) # returns number of element
print(s.ndim) # returns number of dimension
print(s.name) # returns series name
print(s.empty) # returns true oe false check whether series is empty or not

#accessing elements in series
print(s["A"]) #acces by label
# print(s[0])  # acces by position 
#use loc label based indexing safer
print(s.loc["C"]) # uses custom index 
print(s.iloc[2]) #uses integer position
print(s.iloc[[2,3]]) #uses integer position
#slicing - same as numpy
print(s.iloc[0:4])
print(s.loc["B":"D"]) #D - included
#updating
s.loc["A"]=300
print(s) 
#adding element
s["F"]=400
print(s)
#delete and returns a new series by default
s=s.drop("C")
print(s)
#checking if index exist
print("A" in s)
#boolean filtering
print(s[s>2])
print(s)
print( 300 in s.values)
print("A" in s.index)




```

```python
#arithemetic 
s1=pd.Series([1,2,3], index=["A","B","C"])
s2=pd.Series([4,5,6], index=["A","B","G"])
print(s1+s2) # aligns data by index labels
print(s1>4)
#boolean filtering
print(s1[(s1>1) & (s1<3)])
```

```python
#stastical functions
marks=pd.Series([1,2,3,4,5,5])
print(marks.sum())
print(marks.mean())
 
print(marks.mean())
print(marks.mode())
print(marks.median())
print(marks.std())
print(marks.var())
print(marks.unique())
print(marks.nunique()) #returns the number of unique values
print(marks.value_counts()) # return frequency count of each value
print(marks)
```

```python
s=pd.Series([1,0,3], index=["D","B","C"])
print(s)
print(s.sort_values()) # sorting with values
print(s.sort_values(ascending=False)) #sorting in descending order
print(s.sort_index()) #sorting in index
print(s.astype(float))
print(s.apply(str)) #let you apply function to every element
```

```python
marks=pd.Series([1,np.nan,3,np.nan,5,5])
print(marks)
print(marks.isna())
print(marks.isnull())
print(marks.notna())
print(marks.dropna())
print(marks.fillna(8))
```
